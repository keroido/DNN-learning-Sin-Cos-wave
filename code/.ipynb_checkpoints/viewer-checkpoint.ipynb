```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import os
import glob
```

```python
df99 = pd.read_csv('../output/result_00_2_10.csv')
```

```python
x = df99['x']
sin_label = df99['sin(label)']
cos_label = df99['cos(label)']
sin_pre = df99['sin(prediction)']
cos_pre = df99['cos(prediction)']
fig, (axL, axR) = plt.subplots(ncols=2, figsize=(10,4))

axL.scatter(x, sin_label, linewidth=0.1)
axL.scatter(x, sin_pre, c='r', linewidth=0.1)
axL.set_title('sin')
axL.grid(True)

axR.scatter(x, cos_label, linewidth=0.1)
axR.scatter(x, cos_pre, c='r', linewidth=0.1)
axR.set_title('cos')
axR.grid(True)

fig.show()
```
